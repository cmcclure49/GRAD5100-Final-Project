```python
#functions to make it run easier
def run_sql(sql, engine):
    with engine.connect() as conn:
        result = conn.execute(sqla.text(sql))
    return result.all()


def df_select(sql, engine):
    with engine.connect() as conn:
        result = pd.read_sql(sqla.text(sql), con=conn)
    print(f"Retrieved {result.shape[0]} records")
    return result
```

```python
#Problem 4 (Making a winner to judge table) 
winners_to_judge = df_select("""
    SELECT chefdetails.name, chefdetails.season, IF(judges.guestJudge IS NOT NULL, 'Yes', 'No') as was_a_guest_judge
    FROM chefdetails
    LEFT JOIN judges ON chefdetails.name = judges.guestJudge AND chefdetails.season = judges.season
    WHERE chefdetails.placement = 1
""", engine)

winners_to_judge
```

```python
#Problem #5 (Making a loser to judge table)  
losers_to_judge = df_select("""
    SELECT chefdetails.name, chefdetails.season, IF(judges.guestJudge IS NOT NULL, 'Yes', 'No') as was_a_guest_judge
    FROM chefdetails
    INNER JOIN judges ON chefdetails.name = judges.guestJudge AND chefdetails.season = judges.season
    WHERE chefdetails.placement != 1
""", engine)

losers_to_judge
```
